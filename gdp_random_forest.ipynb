# Create ML model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data
```

```python
# Split data
X = data[['x1', 'x2', 'x3', 'x4', 'x5']].values
y = data['y'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
# Train the model
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
# Make predictions
# new_data = y0
# ,2.456,4.00,1386.96,6.4,406426
# 104.7474,2.837,4.04,1462.59,4.4,405948

new_data = [[2.456,4.00,1386.96,6.4,406426], [2.837,4.04,1462.59,4.4,405948]]
new_data_scaled = scaler.transform(new_data)
predictions = model.predict(new_data_scaled)
print(predictions)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Predict the GDP values for the test data
y_pred = model.predict(X_test)

# Calculate the performance metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the performance metrics
print('R^2 score:', r2)
print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)

# Visualize the predicted vs actual values
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, c='orange')
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'b--', lw=2)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
plt.show()
```

```python


# Train and evaluate models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
}

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    results.append([name, r2, mae, rmse])
    
# Convert results to a DataFrame
results_df = pd.DataFrame(results, columns=['Model', 'R^2', 'MAE', 'RMSE'])

# Visualize results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# R^2 plot
axs[0].bar(results_df['Model'], results_df['R^2'], color='blue')
axs[0].set_title('R^2 Score')
axs[0].set_ylim(0, 1)

# MAE plot
axs[1].bar(results_df['Model'], results_df['MAE'], color='red')
axs[1].set_title('Mean Absolute Error')

# RMSE plot
axs[2].bar(results_df['Model'], results_df['RMSE'], color='green')
axs[2].set_title('Root Mean Squared Error')

# Show the plots
plt.tight_layout()
plt.show()
```
