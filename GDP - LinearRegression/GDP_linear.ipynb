# Create ML model

## Linear Regression - feature X5

```python
data[["x5"]]
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, x_test, y_train, y_test = train_test_split(data.x5, data.y, test_size=0.2, random_state=0)
print("completed")
```

```python
data.plot(kind="scatter", x="x5", y="y")
plt.show()
```

```python
plt.scatter(X_train, y_train, label="training data", color="b", alpha=0.7)
plt.scatter(x_test, y_test, label="testing data", color = "g", alpha=0.7)
plt.legend()
plt.title("test train split")
plt.show()
```

```python
# Create linear model
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
#  Make prediction
prediction = LR.predict(x_test.values.reshape(-1,1))
plt.plot(x_test, prediction, label="LinRegression", color="b")
plt.scatter(x_test, y_test, label="Actual test data", color = "g", alpha=0.7)
plt.legend()
plt.show()
```

```python
#  predict value of specific x5

LR.predict(np.array([[239368]]))[0]
```

## Linear Regression for 2 dimensional problem - x3 * x5 ,  x1 * x5


```python
data["x3x5"] = data["x3"] * data["x5"]
data["x1x5"] = data["x1"] * data["x5"]

data.plot(kind="scatter", x="x1x5", y="y")
plt.show()
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, x_test, y_train, y_test = train_test_split(data.x3x5, data.y, test_size=0.2, random_state=0)
print("completed")

plt.scatter(X_train, y_train, label="training data", color="b", alpha=0.7)
plt.scatter(x_test, y_test, label="testing data", color = "g", alpha=0.7)
plt.legend()
plt.title("test train split")
plt.show()

# Create linear model
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)

#  Make prediction
prediction = LR.predict(x_test.values.reshape(-1,1))
plt.plot(x_test, prediction, label="LinRegression", color="b")
plt.scatter(x_test, y_test, label="Actual test data", color = "g", alpha=0.7)
plt.legend()
plt.show()

#  predict value of specific x5

LR.predict(np.array([[239368]]))[0]


```
